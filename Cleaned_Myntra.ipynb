```python
df['Ratings'] = df['Ratings'].fillna(0)
df['Reviews'] = df['Reviews'].fillna(0)
```

```python
df['DiscountPrice (in Rs)'] = df['DiscountPrice (in Rs)'].fillna(df['OriginalPrice (in Rs)'])
```

```python
df['DiscountOffer'] = df['DiscountOffer'].fillna('No Offer')
```

```python
df['discount_percentage'] = (df['OriginalPrice (in Rs)'] - df['DiscountPrice (in Rs)']/df['OriginalPrice (in Rs)']) * 100
```

```python
df['Individual_category'] = df['Individual_category'].str.lower().str.strip()
df['SizeOption'] = df['SizeOption'].str.upper()
```

```python
df['OriginalPrice (in Rs)'] = pd.to_numeric(df['OriginalPrice (in Rs)'],errors='coerce')
```

```python
df['discount_category'] = df['discount_percentage'].apply(lambda x: 'High' if x > 50 else ('Medium' if x > 20 else 'Low'))
```
